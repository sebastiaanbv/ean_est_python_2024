### Ejercicio 4


Escriba el código para categorizar  el total de ingresos  baja, media y alta volatilidad, donde:

`
si totrev > (percentil 75 de totrev para el área dada):
Nivel_totrev = 'Alto'
o si VolStat > (percentil 25 de totrev para el área dada):
Nivel_totrev  = 'Medio'
de lo contrario:
Nivel_totrev  = 'Bajo'`

```python
percentil_75 = df['totrev'].quantile(0.75)
percentil_25 = df['totrev'].quantile(0.25)

# Definir la función para clasificar 'totrev'
def clasificar_totrev(x):
    if x > percentil_75:
        return 'Alto'
    elif x > percentil_25:
        return 'Medio'
    else:
        return 'Bajo'

print(f"Percentil 75 (umbral para 'Alto'): {percentil_75}")
print(f"Valores clasificados como 'Medio' están entre {percentil_25} y {percentil_75}.")
print(f"Percentil 25 (umbral para 'Bajo'): {percentil_25}")

df['Nivel_totrev'] = df['totrev'].apply(clasificar_totrev)
print(df['Nivel_totrev'].value_counts())

# Mostrar el resultado
print("\n")
print(df)
```

### Ejercicio 5

Realice un  boxplot para la variable totmou segmentado por las variable  credict_x y Nivel_totrev

```python
# Crear un listado para almacenar los datos de los segmentos para el boxplot
boxplot_data = []
labels = []
colors = ['lightblue', 'lightgreen', 'lightyellow']

# Generar los segmentos para el boxplot basado en 'creditcd_x' y 'Nivel_totrev'
for credit in df['creditcd'].unique():
    for nivel in df['Nivel_totrev'].unique():
        segment_data = df[(df['creditcd'] == credit) & (df['Nivel_totrev'] == nivel)]['totmou']
        if len(segment_data) > 0:
            boxplot_data.append(segment_data)
            labels.append(f'{credit}-{nivel}')

# Crear el boxplot
fig, ax = plt.subplots()
ax.set_ylabel('totmou')

bplot = ax.boxplot(boxplot_data, patch_artist=True, tick_labels=labels, showfliers=False)

# Añadir colores a los segmentos
for patch, color in zip(bplot['boxes'], colors * len(df['creditcd'].unique())):
    patch.set_facecolor(color)

# Mostrar el gráfico
plt.xticks(rotation=45)
plt.show()
```
